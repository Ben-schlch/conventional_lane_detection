# Exkurs 1: Kamerakalibrierung zur Entzerrung der Bilder
Die vorhandenen Bilder sind aufgrund der Linsen- und Kameraeigenschaften verzerrt. Entzerren Sie die Bilder mithilfe der [Kamerakalibrierungsroutinen von OpenCV](https://docs.opencv.org/4.5.3/dc/dbb/tutorial_py_calibration.html) und den aufgezeichneten Bildern.

```python
# einlesen eines Beispielbildes
img1 = cv.cvtColor(cv.imread('./images/Udacity/calib/calibration1.jpg'), cv.COLOR_BGR2RGB)
plt.imshow(img1)
plt.title('Kalibrierungsbild')
plt.show()
```

```python
GRID_SIZE = (9, 6)
calib_imgs = [cv.cvtColor(cv.imread(f'./images/Udacity/calib/calibration{i}.jpg'), cv.COLOR_BGR2RGB) for i in range(1,21)]
undistorted_imgs = []
t_cap_prev = 0
objpoints = []
imgpoints = []
obj_p = np.zeros((6*9, 3), np.float32)
obj_p[:,:2] = np.mgrid[0:9,0:6].T.reshape(-1,2)


for k in range(len(calib_imgs)):
    calib_img = calib_imgs[k]
    gray_calib_img = cv.cvtColor(calib_img, cv.COLOR_RGB2GRAY)
    
    ret, corners = cv.findChessboardCorners(gray_calib_img, GRID_SIZE, None)
    
    if ret:
        criteria = (cv.TERM_CRITERIA_EPS + cv.TERM_CRITERIA_MAX_ITER, 30, 0.001)
        corners_subpix = cv.cornerSubPix(gray_calib_img, corners=corners, winSize=(11,11), zeroZone=(-1,1), criteria=criteria)
        
        imgpoints.append(corners_subpix)
        objpoints.append(obj_p)
        
    
    # ret, mtx, dist, rvecs, tvecs = cv.calibrateCamera(objpoints, imgpoints, gray_calib_img.shape[::-1], None, None)
    # 
    # h, w = calib_imgs[0].shape[:2]
    # newcameramatrix, roi = cv.getOptimalNewCameraMatrix(mtx, dist, (w,h), 1, (w,h))
    # 
    # # calib_imgs = [cv.undistort(img, mtx, dist, None, newcameramatrix) for img in calib_imgs]
    # undist_img = cv.undistort(calib_imgs[0], mtx, dist, None, newcameramatrix)
    # x, y, w, h = roi
    # undist_img = undist_img[y:y+h, x:x+w]
    # undistorted_imgs.append(undist_img)

plt.imshow(calib_imgs[1])
plt.show()
```

```python
n = 2
gray_calib_img_0 = cv.cvtColor(calib_imgs[n], cv.COLOR_RGB2GRAY)
ret, mtx, dist, rvecs, tvecs = cv.calibrateCamera(objpoints, imgpoints, gray_calib_img_0.shape[::-1], None, None)
h, w = calib_imgs[n].shape[:2]
newcameramatrix, roi = cv.getOptimalNewCameraMatrix(mtx, dist, (w,h), 1, (w,h))
undist_img_0 = cv.undistort(calib_imgs[n], mtx, dist, None, newcameramatrix)
x, y, w, h = roi
undist_img_0 = undist_img_0[y:y+h, x:x+w]

plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.imshow(calib_imgs[n])
plt.subplot(1,2,2)
plt.imshow(undist_img_0)
plt.show()
```
